```python
import pandas as pd
from konlpy.tag import Okt
```

```python
data = pd.read_csv('data.csv')
data
```

## 강의평 5개 미만 삭제

```python
lecture_review = data.groupby('lecture_name').count()
lecture_review = lecture_review.drop(lecture_review.columns[1:], axis=1)

lecture_review = lecture_review[lecture_review['lecture_code']>5]
lecture_review = lecture_review.reset_index()

pre = lecture_review['lecture_name'].unique()

for i in data.index :
    if data['lecture_name'][i] not in pre : 
        data = data.drop([i])
        
data
```

## 특수문자 제거

```python
data['lecture_review'] = data['lecture_review'].str.replace(pat=r'[^\w^ ^.]', repl=r'.', regex=True)

data['lecture_review'] = data['lecture_review'].str.replace(pat=r'[.]+', repl=r'.', regex=True)
data
```

## 문장 쪼개기

```python
data['lecture_review'] = data['lecture_review'].astype(str)

data['sentence'] = data['lecture_review'].str.split(".")

for i in data.index :
    for word in data['sentence'][i] :
        if word == " " :
            data['sentence'][i].remove(word)
        
data
```

## 단어 쪼개기 - okt.pos로 품사태깅 후 조사, 구두점 삭제

```python
okt = Okt()

data['word'] = 0

for i in data.index:
    data['word'][i] = okt.pos(data['lecture_review'][i])

data
```

```python
for i in data.index :
    for word in data['word'][i] :
        if word[1]=='Josa' or word[1]=='Punctuation' or word[1]=='Suffix':
            data['word'][i].remove(word)

data
```

## Nan 값 제거

```python
data = data.dropna()
data = data.reset_index()
data = data.drop(data.columns[0], axis=1)
data
```

## 단어 쪼갠 후 품사 제거

```python
for i in data.index :
    word_result = []
    for word in data['word'][i] :
        word_result.append(word[0])
    data['word'][i] = word_result

data
```

```python
data.to_excel("clean_data.xlsx")
```
