```python
import pandas as pd

data = pd.read_csv('결과_total4.csv')
data = data. drop(['Unnamed: 0'], axis=1)
data
```

```python
data2 = pd.read_excel('data_extractive_20.xlsx')
data2 = data2. drop(['Unnamed: 0'], axis=1)
data2
```

```python
result = pd.merge(left=data, right=data2, how='left', on=['lecture_name','professor_name'], sort=False)

for i in result.index : 
    result['extractive'][i] = str(result['extractive'][i]).replace('\n'," ")

result = result.drop(['lecture_code_y'], axis=1)
result
```

```python
result.to_excel('merge_result.xlsx')
```
