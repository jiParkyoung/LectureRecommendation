```python
import pandas as pd

data = pd.read_excel('merge_result.xlsx')
data = data.drop([data.columns[0]], axis=1)
data
```

```python
jeon = data[data['classification1']=='전']
jeon
```

```python
gyeo = data[data['classification1']!='전']
gyeo
```

```python
jeon_info = pd.read_excel('학과.xlsx', header=None)
jeon_info
```

```python
jeon_info.columns = ['classification1', 'classification2']
jeon_info
```

```python
jeon_info['classification1'] = jeon_info['classification1'].astype(str)
jeon_info
```

```python
jeon['classification1'] = ''

for i in jeon.index : 
    jeon['classification1'][i] = jeon['lecture_code_x'][i][:4]
    
jeon
```

```python
jeon['classification1'] = jeon['classification1'].astype(str)
jeon
```

```python
result_jeon = pd.merge(left=jeon, right=jeon_info, how='left', on='classification1', sort=False)
result_jeon
```

```python
result_jeon.to_excel('전공.xlsx')
```
